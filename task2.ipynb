```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

```python
x = iris_df.iloc[:, [0, 1, 2, 3]].values


from sklearn.cluster import KMeans
s = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    s.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, 
plt.plot(range(1, 11),s)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('S') # Within cluster sum of squares
plt.show()
```

```python


kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)

plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 50, c = 'black', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 50, c = 'orange', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s =50, c = 'pink', label = 'Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'purple', label = 'Centroids')

plt.legend()
```

```python
y_kmeans = np.where(y_kmeans==0, 'Iris-setosa', y_kmeans)
y_kmeans = np.where(y_kmeans=='1', 'Iris-versicolour', y_kmeans)
y_kmeans = np.where(y_kmeans=='2', 'Iris-virginica', y_kmeans)
```

```python

data_with_clusters = iris_df.copy()
data_with_clusters["Cluster"] = y_kmeans
print(data_with_clusters.head(5))
```

```python
sns.barplot(x="Cluster",y="petal width (cm)",data=data_with_clusters)
plt.show()
sns.barplot(x="Cluster",y="sepal width (cm)",data=data_with_clusters)
plt.show()
sns.barplot(x="Cluster",y="petal length (cm)",data=data_with_clusters)
plt.show()
sns.barplot(x="Cluster",y="sepal length (cm)",data=data_with_clusters)
plt.show()
```
